```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
```

# base on the shape of scatter plot, we must use  regression :

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = train["Height"].values
Y = train["Weight"].values
```

```python
X
```

```python
Y
```

```python
X.reshape(-1 ,1)
```

```python
Y.reshape(-1 ,1)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , test_size=0.2, random_state=0) 
```

```python
model = LinearRegression()
```

```python
model.fit(X_train.reshape(-1 ,1) , Y_train.reshape(-1 ,1))
```

```python
X_model_predict = model.predict(X_test.reshape(-1 ,1))
```

```python
Y_model_predict = model.predict(Y_test.reshape(-1 ,1))
```

```python
plt.scatter(X_train.reshape(-1,1),Y_train.reshape(-1,1) , color='orange')
plt.plot(X_test.reshape(-1 ,1) , X_model_predict.reshape(-1 , 1) , color = "purple")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
model.score(X.reshape(-1 ,1) , Y.reshape(-1 ,1))
```
