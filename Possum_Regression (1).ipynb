```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df.sex=le.fit_transform(df.sex)
```

```python
X = df.drop(columns = "age")
Y = df["age"]
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_s = sc.fit_transform(X)
```

```python
Y_s = sc.fit_transform(Y.values.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test , Y_train , Y_test = train_test_split(X_s , Y_s , random_state = 2, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
```

```python
model1.fit(X_train , Y_train)
```

```python
model1_p = model1.predict(X_train)
```

```python
model1_score = model1.score(X_s , Y_s)
```

```python
model1_score
```

```python
plt.scatter(Y_train,model1_p, marker = '*' , c = 'purple')
plt.plot([min(Y_train), max(Y_train)], [min(Y_train), max(Y_train)],linestyle='--', color='black', linewidth=2)
plt.xlabel("original Values")
plt.ylabel("Predict")
plt.show()
```

```python
import keras


from keras.models import Sequential
from keras.layers import Conv2D

model2 = Sequential()



model2.add(keras.layers.Dense(units = 32 , activation = 'relu'))

model2.add(keras.layers.Dense(units = 20 , activation = 'relu'))

model2.add(keras.layers.Dense(units = 8 , activation = 'softmax'))



model2.compile(optimizer='sgd', loss='mean_squared_error', metrics=['mae', 'mse'])


model2.fit(X_train , Y_train , batch_size=35, epochs = 80)
```

```python
model2_p = model2.predict(X_train)
```

```python
from keras.utils import plot_model
plot_model(model2, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
model2.summary()
```

```python
x = []
plt.plot(model2_p, linestyle = 'dotted')
for i in range(0,len(Y_test)):
    x.append(i)
plt.scatter(x, Y_test)
plt.show()
```
