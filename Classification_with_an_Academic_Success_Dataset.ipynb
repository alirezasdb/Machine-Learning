```python
test = test.fillna(0, inplace=True)
```

```python
df_finall = pd.concat([train , test])
```

```python
df_finall.fillna(0, inplace=True)
```

```python
Y_unique = df_finall["Target"].value_counts()
plt.bar("Graduate" , Y_unique["Graduate"]),
plt.bar("Dropout" , Y_unique["Dropout"]),
plt.bar("Enrolled" , Y_unique["Enrolled"])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
sc = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split

X = df_finall.drop(columns = ["id" , "Target"] , axis = 1)
Y = df_finall["Target"]
Y = le.fit_transform(Y)
for c in ['Previous qualification (grade)','Course', 'Admission grade', 'Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)', 'Unemployment rate', 'Inflation rate', 'GDP']:
    X[c] = sc.fit_transform(X[c].values.reshape(-1,1))

X_train , X_test , Y_train , Y_test = train_test_split(X, Y , random_state = 2 , test_size = 0.2)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

estimators = [
    ('lgbm', LGBMClassifier(**{'n_estimators': 250, 'learning_rate': 0.08, 'num_leaves': 40, 'min_data_in_leaf': 40, 'min_gain_to_split': 1.3739536396889442, 'bagging_fraction': 0.9, 'bagging_freq': 1, 'feature_fraction': 0.7})),
    ('xgb', XGBClassifier(**{'iterations': 1168, 'learning_rate': 0.08138388948531615, 'depth': 6, 'l2_leaf_reg': 7.808981138938804}))
]
stacking_model = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
```

```python
stacking_model.fit(X, Y)
```

```python
stacking_model_predict = stacking_model.predict(X_test)
```

```python
stacking_model_score = stacking_model.score(X,Y)
```

```python
stacking_model_score
```

```python
model2_predict = model2.predict(X_test)
```

```python
model2_score = model2.score(X,Y)
```

```python
from sklearn.tree import DecisionTreeClassifier
model3 =DecisionTreeClassifier()
```

```python
# Train Decision Tree Classifer
model3 = model3.fit(X_train,Y_train)

#Predict the response for test dataset
model3_pred = model3.predict(X_test)
model3_s = model3.score(X,Y)
```
