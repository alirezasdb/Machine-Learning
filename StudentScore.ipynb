```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression()
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , random_state = 2 , test_size = 0.2)
```

```python
X_train = X_train.values.reshape(-1 ,1)
```

```python
Y_train = Y_train.values.reshape(-1 ,1)
```

```python
model.fit(X_train.reshape(-1 ,1) , Y_train.reshape(-1 ,1))
```

```python
model_prediction = model.predict(X_test.values.reshape(-1 ,1))
```

```python
X = train.Hours
Y = train.Scores
plt.scatter(X , Y , marker = "*" , c = "purple")
plt.plot(X_test.values.reshape(-1 ,1) , model.predict(X_test.values.reshape(-1 ,1)) , c = "red")
```

```python
model.score(X_test.values.reshape(-1 ,1) , Y_test.values.reshape(-1 ,1))
```
