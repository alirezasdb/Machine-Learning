```python
dictionary = []
for j in range (0 , len(df_train_copy)):
  cleaning = re.sub('[^a-zA-Z0-9]' , ' ' , df_train_copy["text"][j])
  cleaning = cleaning.lower()
  cleaning = cleaning.split()
  cleaning = [stem.stem(words) for words in cleaning if not words in stopwords]
  cleaning = " ".join(cleaning)
  dictionary.append(cleaning)
```

```python
dictionary[100]
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import one_hot
```

```python
vocab_size = 8000
len_sentence = 20
```

```python
visualize_hot = [one_hot(words , vocab_size) for words in dictionary]
```

```python
visualize_hot[100]
```

```python
from tensorflow.keras.utils import pad_sequences
padd = pad_sequences(visualize_hot , maxlen = len_sentence , padding = 'pre')
```

```python
padd[100]
```

```python
Y = df_train["target"]
```

```python
Y
```

```python
Y = np.array(Y)
X = np.array(padd)
```

```python
Y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state = 2 , test_size = 0.2)
```

```python
import keras
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
model.add(keras.layers.Dense(units = 32 , activation = "relu"))
model.add(keras.layers.Dense(units = 20 , activation = "relu"))
model.add(keras.layers.Dense(units = 14 , activation = "relu"))
model.add(keras.layers.Dense(units = 11 , activation = "relu"))
model.add(keras.layers.Dense(units = 8 , activation = "softmax"))
```

```python
model.compile(loss = "sparse_categorical_crossentropy" , optimizer = "adam" ,  metrics=["accuracy"])
```

```python
Model = model.fit(X,Y,validation_data=(X_test,Y_test),epochs=80,batch_size=32)
```
